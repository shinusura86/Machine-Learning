```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)

X = np.random.rand(500, 100) * 100

Y = 3.5 * X + np.random.randn(500, 100) * 20
```

```python
X
```

```python
Y
```

```python
plt.figure(figsize=(8,6))

# Select the first feature of X and Y for a 2D plot visualization
X_plot = X[:, 0].reshape(-1, 1) # Reshape for sklearn's LinearRegression
Y_plot = Y[:, 0]

# Re-fit a Linear Regression model using the selected single feature
model = LinearRegression()
model.fit(X_plot, Y_plot)

# Generate predictions for the selected single feature
Y_pred_plot = model.predict(X_plot)

# Scatter plot of the selected data points
plt.scatter(X_plot, Y_plot, color='blue', label='Data Points')

# Plot the regression line
plt.plot(X_plot, Y_pred_plot, color='red', linewidth=2, label='Regression Line')

plt.title('Linear Regression on Random Dataset (First Feature)')
plt.xlabel('X (First Feature)')
plt.ylabel('Y (First Feature)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Slope (Coefficient):", model.coef_[0][0])
print("Intercept:", model.intercept_[0])
```
